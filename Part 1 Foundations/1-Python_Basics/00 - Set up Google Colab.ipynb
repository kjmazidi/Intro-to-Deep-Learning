# Google colab

Google colab is a free online platform where you can run notebooks on Google's cloud servers. Advantages of using Colab:

* you don't have to install libraries on your own machine
* you can get free access to GPUs in the cloud
* it's easy to share notebooks

### Google drive

Colab is integrated with Google drive.  The first time you go to [colab.research.google.com](https://colab.research.google.com), you will be prompted for your Google login. Then you will be asked to specify a folder for your notebooks in Google drive. The nice thing about files in this folder is that you can right-click and open them in Drive, which will open Google colab.


## What is a notebook?

The notebooks you run in colab are a mixture of code and commentary, organized in cells. These are very similar to Jupyter notebooks. In fact Colaboratory is built on top of Jupyter notebook. 

Add code or text cells with the +Code and +Text buttons up top. 

### Text cells

Text cells for commentary use markdown syntax. See the [markdown guide](https://colab.research.google.com/notebooks/markdown_guide.ipynb). The beauty of markdown is that you can format text as you type with simple notation. You can create:

* bulleted lists by starting lines with '* '
* numbered lists by starting lines with '1. '
* headings by hashtags, the more hashtags the smaller the heading
* bold text is surrounded by double asterisks
* italic text is surrounded by single asterisks

You can also add formulas, hyperlinks, and much more. 

To edit a text cell, double-click on it. Exit by clicking outside of it. 


### Code cells

 The play icon runs the code in a cell. You can move any cell up or down with the arrows in the cell at the right of the cell. You will also see icons there to delete the cell, add comments, and more. 

 The Runtime menu at the top has several important features:

 * interrrupt execution - helpful for stopping a cell that keeps running
 * change runtime type - change from standard execution to GPU or TPU
 * keyboard shortcuts for your system


### Things to notice

* There is a share button at the top right. This allows you to easily share your notebooks. 
* There are also menu items at the top with self-explanatory labels like File, Edit, Help.
